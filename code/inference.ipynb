## Testing File

```python
from keras.models import load_model
from skimage.measure import label, regionprops
import cv2
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

### Loading the Preprocesesed Data

```python
# modelMNIST load

model = load_model('modelMNIST.h5')


# load data 
a1 = np.load('d0.npy',allow_pickle=True)
a2 = np.load('d1.npy',allow_pickle=True)
a3 = np.load('d2.npy',allow_pickle=True)
a4 = np.load('d3.npy',allow_pickle=True)
```

### Loading the Training Labels 

```python
lab1 = np.load('../Data/lab0.npy')
lab2 = np.load('../Data/lab1.npy')
lab3 = np.load('../Data/lab2.npy')
```

```python
labfin1 = lab1[:400]
labfin2 = lab2[:200]
labfin3 = lab3[:200]

finalLabels = np.concatenate((labfin1,labfin2,labfin3),axis=0)
```

```python
#concatenate data
data = np.concatenate((a1,a2,a3,a4),axis=0)
```

### Final Evaluation for 800 Images

```python

y_train = finalLabels
accuracy = 0


for i in range(len(data)):
    sum = 0
    for j in range(len(data[i])):
        img = data[i][j]
        img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        img = img.reshape(1, 28, 28, 1)
        img = img.astype('float32')
        img /= 255
        pred = model.predict(img)
        sum = sum + np.argmax(pred)
    if sum == y_train[i]:
        accuracy += 1
    print(sum, y_train[i])

print("Accuracy: ", accuracy/len(y_train))
```
